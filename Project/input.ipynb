```python
# import output.csv as a dataframe

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# read in the csv file
df = pd.read_csv('output (1).csv')

df.head()
```

```python
df['duration'] = pd.to_datetime(df['dtend']) - pd.to_datetime(df['dtstart'])
# get duration in minutes
df['duration'] = df['duration'].dt.total_seconds() / 60
df.head()
```

```python
print(len(df))
```

```python
# only keep the columns summary, duration, dtstart, detend and location
df = df[['summary', 'duration', 'dtstart', 'dtend', 'location']]

df.head()
```

```python
# add a column for author/speaker in the df as empty column
df['author'] = ''
df.head()
```

```python
# extract type from summary, title from summary and author from summary
# type is the one mentioned in []
# author is the one mentioned after last '-'
# title is the one mentioned before last '-' and after ]
for index, row in df.iterrows():
    summary = row['summary']
    type = summary[summary.find('[')+1:summary.find(']')]
    author = summary[summary.rfind('-')+1:]
    title = summary[summary.find(']')+1:summary.rfind('-')]
    df.at[index, 'type'] = type
    df.at[index, 'author'] = author
    df.at[index, 'title'] = title
    
df.head()
```

```python
# drop summary
df = df.drop(columns=['summary'])
df.head()
# save to csv called input

```

```python
# if the author feild contains a term between [], copy the text after ] in the title column and clear author column

for index, row in df.iterrows():
    author = row['author']
    title = row['title']
    if '[' in author:
        term = author[author.find('[')+1:author.find(']')]
        # title is the text after ']' in the title column
        # print(author[author.find(']')+1:])
        df.at[index,'title'] = author[author.find(']')+1:]
        df.at[index, 'author'] = ''



df.head()
```

```python
df.to_csv('input.csv', index=False)
```
